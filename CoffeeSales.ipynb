```python
df1 = spark.read.csv("/Volumes/workspace/brk/brk/Coffe_sales.csv", header=True, inferSchema=True)
df1.show()
```

```python
df1.printSchema()
```

```python
from pyspark.sql.functions import to_date, date_format, col, sum as spark_sum

# Ensure 'Date' column is in date format and extract 'Year-Month'
df1 = df1.withColumn(
    "YearMonth",
    date_format(
        to_date(col("Date")),
        "yyyy-MM"
    )
)

# Aggregate sales by coffee_name and month using the correct column name
monthly_sales_trend = df1.groupBy(
    "coffee_name",
    "YearMonth"
).agg(
    spark_sum("money").alias("TotalSales")
).orderBy(
    "coffee_name",
    "YearMonth"
)

display(monthly_sales_trend)
```

```python
monthly_sales_trend.plot.area(x="YearMonth", y="TotalSales", color="coffee_name")
```

```python
from pyspark.sql import functions as F

pivoted = monthly_sales_trend.groupBy("coffee_name").pivot("YearMonth").agg(F.first("TotalSales"))
display(pivoted)
```
