```python
data = [
    ("Sales", 3),
    ("Finance", 4),
    ("Marketing", 2)
]
columns = ["department", "count"]
df = spark.createDataFrame(data, columns)

result_df = df.withColumn(
    "repeat_dept",
    F.expr("array_repeat(department, int(count))")
).select(
    F.explode("repeat_dept").alias("department")
)

display(result_df)
```
