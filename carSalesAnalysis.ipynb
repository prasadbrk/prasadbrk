```python
car_sales_df = spark.table(
    "brk.car_sales"
)
display(
    car_sales_df
)
```

```python
car_sales_df.cache()
```

```python
from pyspark.sql import functions as f
ann_income_gender = car_sales_df.groupBy("Gender").agg(f.round(f.avg("Annual Income"),2).alias("Avg_annual_Income"))
display(ann_income_gender)
```

```python
models_by_company=car_sales_df.groupBy("Company",).agg(f.count_distinct("Model").alias("Models")).sort("Models", ascending=False)
display(models_by_company)
```

```python
car_sales_df= car_sales_df.withColumn("Year", f.year("Date")).withColumn("Month", f.month("Date"))
display(car_sales_df)
```

```python
sales_by_yr_mth = car_sales_df.groupBy("Year", "Month").agg(f.sum("Price ($)").alias("Total_Sales")).sort("Year", "Month")
display(sales_by_yr_mth)
```

```python
sales_by_yr_mth_pivot = sales_by_yr_mth.groupBy("Year").pivot("Month").sum("Total_Sales")
display(sales_by_yr_mth_pivot)
```
